# Week 1 Ops Analyzer

This notebook analyzes operational efficiency for three fuel depots.

We'll:
1. Define a KPI function to calculate efficiency percentage.
2. Build a status classifier (Critical / Warning / Normal).
3. Model three sites with their actual and target outputs.
4. Loop through them and print a formatted report.

All logic is explained in these Markdown cells – written for a junior analyst.

```python
def calculate_efficiency(actual_output, target_output):
    """
    Calculate efficiency percentage.
    
    Parameters:
        actual_output (float/int): actual output achieved
        target_output (float/int): target output expected
        
    Returns:
        float: efficiency percentage; returns 0.0 if target is zero to avoid division error.
    """
    if target_output == 0:
        return 0.0
    return (actual_output / target_output) * 100
```

## 1. KPI Function

`calculate_efficiency(actual_output, target_output)` returns the efficiency percentage.

**Formula:**  
Efficiency = (actual_output / target_output) × 100

We check if `target_output` is zero – if it is, we return `0.0` to avoid a division error.

```python
def get_operational_status(efficiency):
    """
    Classify operational status based on efficiency percentage.
    
    Parameters:
        efficiency (float): efficiency percentage
        
    Returns:
        str: 'Critical', 'Warning', or 'Normal'
    """
    if efficiency < 70:
        return "Critical"
    elif efficiency < 90:   # 70 <= efficiency < 90
        return "Warning"
    else:
        return "Normal"
```

## 2. Status Classifier

`get_operational_status(efficiency)` returns:

- **"Critical"** – if efficiency is less than 70%
- **"Warning"** – if efficiency is between 70% and 89%
- **"Normal"** – if efficiency is 90% or higher

This helps operations managers quickly identify which sites need immediate attention.

```python
# Data for three fuel depots
sites = [
    {"name": "North Depot", "actual_output": 8200, "target_output": 10000},
    {"name": "South Depot", "actual_output": 6500, "target_output": 8000},
    {"name": "East Depot",  "actual_output": 9500, "target_output": 10000}
]
```

## 3. Data Modeling

We model **three fuel depots** as a list of dictionaries.

Each site has:
- `name` – site identifier
- `actual_output` – actual throughput (e.g., litres delivered)
- `target_output` – planned throughput

Using a list makes it easy to add more sites in the future.

```python
# Print a header
print("=" * 60)
print("OPERATIONAL EFFICIENCY REPORT")
print("=" * 60)
print()

# Loop through each site
for site in sites:
    name = site["name"]
    actual = site["actual_output"]
    target = site["target_output"]
    
    # Calculate efficiency
    efficiency = calculate_efficiency(actual, target)
    
    # Determine status
    status = get_operational_status(efficiency)
    
    # Print formatted report
    print(f"Site: {name}")
    print(f"  Actual Output: {actual}")
    print(f"  Target Output: {target}")
    print(f"  Efficiency: {efficiency:.1f}%")
    print(f"  Status: {status}")
    print("-" * 40)

print("End of report.")
```

This notebook demonstrates:
- A reusable KPI function for efficiency calculations.
- A clear status classification based on thresholds.
- A scalable data structure (list of site dictionaries) that can be extended.
- A simple loop that generates an operational report.
